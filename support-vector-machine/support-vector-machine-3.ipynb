```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC 
trainset_input = pd.read_csv('Training_DB_Multifont inputs.csv', header = None)
testset_input = pd.read_csv('Testing_DB_Multifont inputs.csv', header = None)
trainset_output= pd.read_csv('Training_DB_Multifont outputs.csv', header = None)
testset_output = pd.read_csv('Testing_DB_Multifont oututs.csv', header = None)
trainset_output = trainset_output.idxmax(axis=1)
testset_output = testset_output.idxmax(axis=1)
```

```python
report = pd.DataFrame(np.zeros((90,6)), columns=['kernel','degree','c_searchline', 'num_of_vectors','score_train', 'score_test'])
c_searchline = [.015, .2 , 2, 12 ]
Kernels =['poly','linear', 'rbf', 'sigmoid']
degrees =[2,3,4,5]
index = 0 

for c in c_searchline :
    
    for k in Kernels :
       
        
        for deg in degrees:
            
            model = SVC(gamma='auto', C = c, kernel = k , degree = deg , verbose = True , decision_function_shape = 'ovr')
            index+=1
               
        
    
            model.fit(trainset_input ,trainset_output)
            nSV= len(model.support_)
            score_train =  model.score(trainset_input ,trainset_output)
            score_test =  model.score(testset_input,testset_output)
            report.iloc[index]= [k,deg,c,nSV, score_train , score_test]
```

```python
outcome= pd.read_csv("report.csv")
```

```python
def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]
```

```python
display(outcome.style.apply(highlight_max, subset=['score_train','score_test']))
```

```python
from sklearn.metrics import confusion_matrix
best_model = SVC(C=1 , kernel ='linear', degree= 2, gamma='auto',decision_function_shape='ovr')
best_model.fit(trainset_input ,trainset_output)
testset_prediction = best_model.predict(testset_input)
pd.set_option('display.max_columns', 30)
display(pd.DataFrame(confusion_matrix(testset_output,testset_prediction)))
```
